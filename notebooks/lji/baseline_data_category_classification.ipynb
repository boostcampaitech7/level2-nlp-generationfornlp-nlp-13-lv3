```python
import pandas as pd
import re

# 데이터 로드
sample_df = pd.read_csv('sample_data.csv')
sample_df12 = sample_df.copy()

# 지정된 카테고리 목록
valid_categories = [
    '국어-독서', '국어-문학', '국어-화법과작문', '국어-언어', '국어-매체',
    '사회-생활과윤리', '사회-윤리와사상', '사회-한국지리', '사회-세계지리',
    '사회-동아시아사', '사회-세계사', '사회-경제', '사회-정치와법', '사회-문화'
]
```

```python
# 카테고리 열에서 숫자와 점 제거 (예: '1.', '2.' 등)
sample_df12['cleaned_category'] = sample_df12['category'].str.replace(r'^\d+\.\s*', '', regex=True)

# 지정된 카테고리에 해당하는 데이터 추출
standard_categories = sample_df12[sample_df12['cleaned_category'].isin(valid_categories)]

# 지정된 카테고리에 해당하지 않는 데이터 추출
non_standard_categories = sample_df12[~sample_df12['cleaned_category'].isin(valid_categories)]
```

```python
def clean_non_standard_category(category):
    # 줄바꿈 문자 제거 및 공백 정리
    category = re.sub(r'\s+', ' ', category).strip()
    
    # 접두어 제거
    prefixes_to_remove = ['과목:', '세부 주제:', '결과:', '카테고리:']
    for prefix in prefixes_to_remove:
        category = category.replace(prefix, '').strip()
    
    # 유사한 카테고리 통합 및 기타 정제 작업
    if category in ['심리-심리학', '과학-심리학']:
        return '과학-심리학'
    if category in ['사회-범죄와법', '사회-범죄와 법']:
        return '사회-범죄'
    if category in ['자연-생물학', '과학-생물', '과학-생물학']:
        return '과학-생물학'
    
    # 기타 정제 작업
    category = category.replace('사회-노동과 경제', '사회-경제')
    
    return category
```

```python
# 비표준 카테고리에 대한 정제 함수 적용
non_standard_categories['cleaned_category'] = non_standard_categories['cleaned_category'].apply(clean_non_standard_category)

# None 값을 가진 행 제거
non_standard_categories_cleaned = non_standard_categories.dropna(subset=['cleaned_category'])
```

```python
import re

def is_standard_category(category):
    return re.match(r'^(사회|수학|과학|국어)-', category) is not None

# 표준 카테고리 추출
ss_standard_categories = non_standard_categories_cleaned[non_standard_categories_cleaned['cleaned_category'].apply(is_standard_category)]

# 비표준 카테고리 추출
non_standard_categories = non_standard_categories_cleaned[~non_standard_categories_cleaned['cleaned_category'].apply(is_standard_category)]
```

```python
# id에서 숫자 부분만 추출하여 새로운 열에 저장
combined_df['numeric_id'] = combined_df['id'].apply(lambda x: int(x.split('-')[-1]))

# numeric_id를 기준으로 정렬
combined_df_sorted = combined_df.sort_values(by='numeric_id')
```

```python
# cleaned_category 열의 값 개수 계산
category_counts = combined_df_sorted['cleaned_category'].value_counts()

# 전체 개수 계산
total_count = len(combined_df_sorted)

# 비율 계산
category_percentages = (category_counts / total_count) * 100

# 결과 출력
print("각 카테고리의 비율:")
for category, percentage in category_percentages.items():
    print(f"{category}: {percentage:.2f}%")
```
